```python
%pip install seaborn 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%pip install openpyxl
```

```python
df=pd.read_excel('emp_data.xlsx')
```

```python
df=pd.DataFrame(df)
df
```

```python
df.info()
```

```python
# Count of employees per department

plt.figure(figsize=(6, 3))
sns.countplot(data=df, x="department", order=df["department"].value_counts().index,color='brown')
plt.xticks(rotation=45)
plt.title("Employee Count by Department")
plt.show()
```

```python
#Employee counts by Length of Services

sns.histplot(df['length_of_service'],bins=20,kde=True,color="green")
plt.title("Counts Of Employee By Length of services")
plt.ylabel("Count Of Employees")
plt.show()
```

```python
#Awards Won by Departments

awd=df[df['awards_won']==1]
awards_won_dept=awd.groupby('department')['awards_won'].sum().plot(kind='pie')
plt.title("Awards won by Deparments")
```

```python
#Gender Distribution based on KPIs met more than 80


plt.figure(figsize=(6,5))
kpi=df[df['KPIs_met_more_than_80']==1]
gender_kpi=kpi.groupby('gender')['KPIs_met_more_than_80'].count().plot(kind='barh')
plt.xlabel('Count of KPIs_met_more_than_80')
plt.title("KPIs_met_more_than_80 by Gender Distribution")
plt.show()

```

```python
# Average training score per department

dept_training = df.groupby("department")["avg_training_score"].mean().sort_values()
plt.figure(figsize=(6, 4))
sns.barplot(x=dept_training.index, y=dept_training.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Average Training Score by Department")
plt.show()
```

```python
# Distribution of Age

plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=20, kde=True, color="Purple")
plt.title("Age Distribution")
plt.show()
```
